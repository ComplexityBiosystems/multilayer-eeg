```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import matplotlib as mpl

mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
```

```python
from scipy.stats import ks_2samp, ttest_ind, median_test
```

```python
palette = {
    "Control": "0.8",
    "Bipolar": "#8ad5ff",
    "Psychotic": "#feb3ff"
}
```

```python
def plot_measure(ssdf, network_type, measure, ax):
    _hue_order = ["Control", "Bipolar", "Psychotic"]
    hue_order = [x for x in _hue_order if x in ssdf.merged_patient_type.values]

    sns.boxplot(
        data=ssdf,
        x="band",
        y="value",
        hue="merged_patient_type",
        hue_order=hue_order,
        palette=palette,
        saturation=1,
        fliersize=0,
        ax=ax
    )
    sns.stripplot(
        data=ssdf,
        x="band",
        y="value",
        hue="merged_patient_type",
        hue_order=hue_order,
        color="0.3",
        dodge=True,
        ax=ax
    )
    ax.legend(loc=(1, 0), frameon=False)
    handles, labels = ax.get_legend_handles_labels()
    new_labels = [
        label_simplifier[x]
        for x in labels
    ]
    ax.legend(handles, new_labels, loc=(1, 0), frameon=False)
    
    ax.set_xlabel("Band")
    ax.set_ylabel(measure)
    ax.set_title(network_type.replace("_", " "))
```

```python
def get_pvals(ssdf, network_type, measure):
    pvals_dict = {}
    for band, sssdf in ssdf.groupby("band"):
        _control = sssdf.loc[sssdf.merged_patient_type == "Control", "value"].values
        _bipolar = sssdf.loc[sssdf.merged_patient_type == "Bipolar", "value"].values
        _psychotic = sssdf.loc[sssdf.merged_patient_type == "Psychotic", "value"].values

        pval_bipolar = ttest_ind(_control, _bipolar, equal_var=False).pvalue
        pval_psychotic = ttest_ind(_control, _psychotic, equal_var=False).pvalue

        pvals_subdict = {
            "bipolar_vs_ctrl": pval_bipolar,
            "psychotic_vs_ctrl": pval_psychotic
        }
        pvals_dict[band] = pvals_subdict
    
    pvals_df = pd.DataFrame(pvals_dict)
    pvals_df.columns.name="band"
    pvals_df = pvals_df.T.reset_index()

    pvals_df["measure"] = measure
    pvals_df["network_type"] = network_type
    return pvals_df
```

```python
def _add_bracket(ax, x, ymin, ymax, dx=0.27, yrel = 0.99, dyrel = 0.02, kind=0):
    """
    kind = 0 for short one, 1 for long one
    """
    y = ymin + yrel * (ymax - ymin)
    dy = dyrel * (ymax - ymin)
    ax.plot(
        *np.array([
        [x - dx, y - dy],
        [x - dx, y],
        [x + kind * dx, y],
        [x + kind * dx, y - dy]
        ]).T,
        c="0.2"   
    )
    xm = 0.5 * (x - dx + x + kind * dx)
    ax.text(xm, y, "*", ha="center")
    
def add_pvalue_stars_merged(ax=None, pvals=None, threshold=0.05, correction=None):
    if correction is not None:
        raise NotImplementedError
    
    ymin, ymax =ax.get_ylim()
    for x in pvals.loc[pvals["bipolar_vs_ctrl"] < threshold, "band"].values:
        _add_bracket(ax=ax, x=x, ymin=ymin, ymax=ymax, kind=0, yrel=0.97)
    for x in pvals.loc[pvals["psychotic_vs_ctrl"] < threshold, "band"].values:
        _add_bracket(ax=ax, x=x, ymin=ymin, ymax=ymax, kind=1, yrel = 1.02)

def add_pvalue_stars(ax=None, pvals=None, threshold=0.05, correction=None):
    if correction is not None:
        raise NotImplementedError
    
    dx = 0.18
    ymin, ymax =ax.get_ylim()
    for x in pvals.loc[pvals["bipolar_vs_ctrl"] < threshold, "band"].values:
        _add_bracket(ax=ax, x=x, ymin=ymin, ymax=ymax, kind=1, yrel=1, dx=dx)
    for x in pvals.loc[pvals["psychotic_vs_ctrl"] < threshold, "band"].values:
        _add_bracket(ax=ax, x=x, ymin=ymin, ymax=ymax, kind=1, yrel = 1, dx=dx)
  
```

### Load the data

```python
df = pd.read_pickle("../data/network-measures/network-measures-all-patients.p")
```

### Simplify labels

```python
label_simplifier = {
    "Control": "Control",
    "Bipolar": "BD",
    "Psychotic": "FEP",
    "merged_patient_type": "merged_patient_type"
}
```

### Plot boxplots

```python
pvals_list = []
for measure, sdf in df.groupby("measure_name"):
    fig, axx = plt.subplots(1, 2, figsize=(18, 4), tight_layout=True)
    for ax, (network_type, ssdf) in zip(axx[::-1], sdf.groupby("network_type")):
        plot_measure(ssdf, network_type, measure, ax)
        pvals = get_pvals(ssdf, network_type, measure)
        pvals_list.append(pvals)
        # add the stars
        add_pvalue_stars_merged(ax=ax, pvals=pvals, threshold=0.05)

    filename = f'{measure.lower().replace(" ", "_")}.png'
    fig.savefig("../figures/boxplots/" + filename, dpi=150)

pvals_df = pd.concat(pvals_list)
pvals_df.to_pickle("../figures/boxplots/pvalues.p")
pvals_df.to_excel("../figures/boxplots/pvalues.xlsx")
```

```python
for measure_name in [
    'Average Clustering',
    'Average Shortest Path Length',
    'Betweennes Centrality',
    'Number of connected components',
    'PO Presence',
    'PO vs rest Presence'
]:
    sdf = df.loc[df.network_type == "Fixed_Th"]
    table = sdf.pivot_table(index="patient_code", columns="measure_name", values="value")
    table["merged_patient_type"] = df[["patient_code", "merged_patient_type"]].drop_duplicates().set_index("patient_code")

    fig, ax = plt.subplots(1, 1, figsize=(5, 3.5), tight_layout=True)
    sns.scatterplot(
        data=table,
        x="Density",
        y=measure_name,
        hue="merged_patient_type",
        hue_order = ["Control", "Bipolar", "Psychotic"],
        palette=palette,
        edgecolor="0.2",

    )
    if measure_name ==  'Betweennes Centrality':
        ax.set_ylim(0.009, 0.03)
    legend = ax.legend(frameon=False, loc=(1, 0))
    legend.texts[0].set_text("")
    handles, labels = ax.get_legend_handles_labels()
    new_labels = [
        label_simplifier[x]
        for x in labels
    ]
    legend = ax.legend(handles, new_labels, loc=(1, 0), frameon=False)
    legend.texts[0].set_text("")
    
    filename = f'{measure_name.lower().replace(" ", "_")}_vs_density.png'
    fig.savefig("../figures/scatterplots/" + filename, dpi=150)
```
